```python
def add(a, b): return a + b
def sub(a, b): return a - b
def mul(a, b): return a * b
def div(a, b): return a / b if b != 0 else "Error: Division by zero"

operations = {
    '1': ('Addition', add),
    '2': ('Subtraction', sub),
    '3': ('Multiplication', mul),
    '4': ('Division', div),
    '+': ('Addition', add),
    '-': ('Subtraction', sub),
    '*': ('Multiplication', mul),
    '/': ('Division', div)
}

print("-- Welcome to Calculator App --")

while True:
    try:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        
        print("\nOperations:")
        for key, (name, _) in operations.items():
            if key.isdigit():  # Only print once
                print(f"{key}. {name}")

        op = input("Enter your choice (number or symbol): ")
        if op in operations:
            name, func = operations[op]
            print(f"{name} result: {func(a, b)}")
        else:
            print("Invalid choice.")
    except ValueError:
        print("Please enter valid numeric inputs.")
    
    if input("\nDo you want to continue? (yes/no): ").lower() == "no":
        break

print("Thank you for using the Calculator App!")
```
